```python
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
import time
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv('data.csv', index_col=0)
```

```python
y = df.diagnosis
list = ['id', 'diagnosis']
x = df.drop(list, axis=1)
x.head

oversample = SMOTE()
x_n, y_n = oversample.fit_resample(x,y)
```

```python
data_dia = y_n
data = x_n
data_n_2 = (data - data.mean()) / (data.std())
#data = pd.concat([y, data_n_2], axis=1)
```

```python
components_set = np.concatenate([np.array([1, 3, 5, 10]), np.arange(50,2050,50)])
kernel_set = np.array(['linear', 'rbf','sigmoid'])
```

```python
for i in range(0, components_set.shape[0]):
    for j in range(0, kernel_set.shape[0]):
        
        components = components_set[i]
        kernel = kernel_set[j]
        print(components,kernel)
        #X = pd.read_csv('X_'+str(components_set[i])+'.csv', index_col = 0)
        startTime = time.time() 
        #X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
        X_train, X_test, y_train, y_test = train_test_split(data_n_2, y_n, train_size = 0.7, test_size = 0.3, random_state = 1)
        clf = svm.SVC(kernel = kernel)
        clf.fit(X_train,y_train)
        svm_output = np.zeros(y_train.shape)
        svm_output = clf.predict(X_test)
            
        predicted_output = svm_output
        print(confusion_matrix(y_test, svm_output))
        print(classification_report(y_test, svm_output))
        runTime = time.time() - startTime
        print("T_run:", runTime)
```
